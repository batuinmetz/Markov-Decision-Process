```python
import hiive.mdptoolbox
import numpy as np
import matplotlib.pyplot as plt
import gym
import hiive.mdptoolbox.example
from gym.envs.toy_text.frozen_lake import generate_random_map

#defining the problem
#random_map = generate_random_map(size=4, p=0.9)
random_map = ['SFFF', 'HFFH', 'HHFF', 'FFFG']
env = gym.make('FrozenLake-v1', desc=random_map, is_slippery=True, render_mode="rgb_array")
env.reset()
env_screen = env.render()
#vizualisation
plt.imshow(env_screen)
print(random_map)

#getting the transition and reward matrix
P, R = hiive.mdptoolbox.example.openai("FrozenLake-v1", desc=random_map)
#print(P,R)
#Q-learning
ql = hiive.mdptoolbox.mdp.QLearning(P, R, 0.95, n_iter=100000,alpha_decay=0.99)
ql.run()
print("executed")
```
