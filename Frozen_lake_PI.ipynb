```python
#!pip install gym
#!pip install gym[toy_text]
#!pip install git+https://github.com/hiive/hiivemdptoolbox.git
import numpy as np
import matplotlib.pyplot as plt
import hiive.mdptoolbox
import gym
import hiive.mdptoolbox.example
from gym.envs.toy_text.frozen_lake import generate_random_map
from numpy import cumsum

size_frozen_lake=15
#random_map = generate_random_map(size=size_frozen_lake, p=0.9)
#random_map=['SFHF', 'FFFF', 'FFFF', 'FHFG']
#random_map_6=['SHFFFH', 'FHFFFF', 'FHFFFF', 'FFFFFF', 'FFFFFF', 'FFFFFG']
#random_map_20=['SFFHFFFFFFFFFFFFFFFH', 'FFFFFFFFHFFFFFFFFHFF', 'FHFFFFFFFFFHFHFHFHFF', 'FFHFFHFFFFFFHHFFFFFF', 'FFFFFFFFFFHFFFFFFHHF', 'FFHFHFFFFFFFFHFFFFFF', 'FFFFFHFFFFFFFFFFFFFF', 'FHHFFFFFFHFFFFFFFFFF', 'FFFFFFFFFFFHFHFFFHFF', 'FFFFFFFFFFFFHFFFFFFF', 'HFHFFFFHFHFFFHFFFFFF', 'FFFFFFFFFFFFFFFFFFFF', 'FFFFFFFFFFFFFFFFFFFF', 'FFFFFFHFFFFFFFFFFFFF', 'FFHHFFFFFFHFFFFFFFHF', 'FFFFFFFFFFFFHFFFFFHF', 'FFFFFFFFFFFFFFFFFFFH', 'FFFFFFFFFFFHFHFFFFFF', 'FFFFFFFFFFFFFFFHFHFH', 'FFFFFFFFFFFFFFFFHFFG']
random_map=['SFFFHFFFFFFFFFF', 'FFFFFFFHFFFHFFF', 'FFFHFHFFFFFFFFF', 'FFFFFFFFFFFFFHF', 'FFFFFHHFFHFFFFF', 'FFFFFFFFFFFFFFF', 'FFFFFFHFFHFFHFF', 'FFFFFFFFFFHHFFF', 'FFFFFFFFFFFFFFF', 'FFFFFFFFFFFFFFF', 'FFFFFFFHFFFFFFH', 'FFHFFFFFFFFHFHF', 'FFFFHFFFFFFFHFF', 'FFFHFFFFFFFFFFF', 'FFFFFFHFFHFFFFG']
env = gym.make('FrozenLake-v1', desc=random_map, is_slippery=True, render_mode="rgb_array")
env.reset()
env_screen = env.render()
plt.imshow(env_screen)
print(random_map)
P, R = hiive.mdptoolbox.example.openai("FrozenLake-v1", desc=random_map, is_slippery=True, render_mode="rgb_array")
#print(P,R)
pi = hiive.mdptoolbox.mdp.PolicyIteration(P, R, 0.95)
pi.run()
#print(pi.run())
pi.run_stats
```

```python
time_vector=[]
reward_vector=[]
iteration_vector=[]
for k in range (pi.iter):
    time_vector.append(pi.run_stats[k].get('Time'))
    reward_vector.append(pi.run_stats[k].get('Reward'))
    iteration_vector.append(pi.run_stats[k].get('Iteration'))

plt.plot(iteration_vector, reward_vector, marker='o', markersize=2,color='blue', label='Reward')
plt.xlabel('Number of iterations')
plt.ylabel('Reward')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
#evolution with a parameter gamma tuning: 
gamma_list=[0.80,0.85,0.95,0.99]
color_list=['green','orange','red','black']
for g in range (0,len(gamma_list)):
    pi = hiive.mdptoolbox.mdp.PolicyIteration(P, R, gamma_list[g])
    pi.run()
    reward_vector=[]
    iteration_vector=[]
    for k in range (80):
        reward_vector.append(pi.run_stats[k].get('Reward'))
        iteration_vector.append(pi.run_stats[k].get('Iteration'))
    plt.plot(iteration_vector, reward_vector, marker='o', markersize=2,color=color_list[g], label='R_gamma_'+str(gamma_list[g]))
    plt.xlabel('Number of iterations')
    plt.ylabel('Reward')
    plt.title('Reward vs number of iterations in Frozen Lake with PI')
    plt.legend(loc='lower right')
    plt.grid()
plt.show()
```

```python
#evolution with a parameter gamma tuning: 
gamma_list=[0.80,0.85,0.95,0.99]
color_list=['green','orange','red','black']
for g in range (0,len(gamma_list)):
    pi = hiive.mdptoolbox.mdp.PolicyIteration(P, R, gamma_list[g])
    pi.run()
    time_vector=[]
    iteration_vector=[]
    for k in range (100):
        time_vector.append(pi.run_stats[k].get('Time'))
        iteration_vector.append(pi.run_stats[k].get('Iteration'))
    plt.plot(iteration_vector, cumsum(time_vector), marker='o', markersize=2,color=color_list[g], label='Time'+str(gamma_list[g]))
    plt.xlabel('Number of iterations')
    plt.title('Time vs number of iterations in Frozen Lake with PI')
    plt.ylabel('Time')
    plt.legend(loc='lower right')
    plt.grid()
plt.show()
```

```python
policy_reshape=np.reshape(pi.policy,(size_frozen_lake,size_frozen_lake))
print(policy_reshape)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, xlim=(0, policy_reshape.shape[1]), ylim=(0, policy_reshape.shape[0]))
color_map=['blue','green','orange','red']

for i in range(policy_reshape.shape[0]):
    for j in range(policy_reshape.shape[1]):
        state_corner_1=policy_reshape.shape[0]-i-1
        state_corner_2=j
        state_rectangle = plt.Rectangle([state_corner_2, state_corner_1], 1, 1)
        state_rectangle.set_facecolor(color_map[policy_reshape[i,j]])
        ax.add_patch(state_rectangle)
        
        if policy_reshape[i,j]==0:
            plt.arrow(state_corner_2+0.75, state_corner_1+0.5, -0.25, 0,head_width = 0.2,width = 0.05,ec ='white')
        if policy_reshape[i,j]==1:
            plt.arrow(state_corner_2+0.5, state_corner_1+0.75, 0, -0.25,head_width = 0.2,width = 0.05,ec ='white')
        if policy_reshape[i,j]==2:
            plt.arrow(state_corner_2+0.25, state_corner_1+0.5, 0.25, 0,head_width = 0.2,width = 0.05,ec ='white')
        if policy_reshape[i,j]==3:
            plt.arrow(state_corner_2+0.5, state_corner_1+0.25, 0, 0.25,head_width = 0.2,width = 0.05,ec ='white')
        
plt.axis('on')
plt.title('Best Policy after Policy Iteration')
plt.xlim((0, policy_reshape.shape[1]))
plt.ylim((0, policy_reshape.shape[0]))
plt.tight_layout()
#plt.savefig(title+str('.png'))
plt.show()
```

```python
#evolution with a parameter gamma tuning: 
gamma_list=[0.80,0.85,0.95,0.99]
color_list=['green','orange','red','black']
for g in range (0,len(gamma_list)):
    pi = hiive.mdptoolbox.mdp.PolicyIteration(P, R, gamma_list[g])
    pi.run()
    time_vector=[]
    reward_vector=[]
    for k in range (1000):
        time_vector.append(pi.run_stats[k].get('Time'))
        reward_vector.append(pi.run_stats[k].get('Reward'))
    plt.plot(cumsum(time_vector), reward_vector, marker='o', markersize=2,color=color_list[g], label='Reward'+str(gamma_list[g]))
    plt.title('Reward vs Time in Frozen Lake with PI')
    plt.xlabel('Time')
    plt.ylabel('Reward')
    plt.legend(loc='lower right')
    plt.grid()
plt.show()
```
